```python
from flask import Flask, render_template, request,jsonify
from flask_cors import CORS,cross_origin
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
flipkart_url = "https://www.flipkart.com/search?q=" + 'iphone7'
flipkart_url
```

```python
#uReq is alias of urlopen
#we are asking to open the url
uClient = uReq(flipkart_url)
#uClient gives http response
```

```python
#reading the html page
flipkartPage = uClient.read()
flipkartPage
#extracts all data in the page
```

```python
#html is not readable and not in useful format
#segregate data of different iphones
#for this we have to click on the link of one particular iphone and read it
#we close the url client
uClient.close()
```

```python
#beautiful soup removes all unnecesart text and information and create some structure
#bs4 works for bith html parser and xml parser
flipkart_html = bs(flipkartPage,'html.parser')
```

```python
#div class can be got from inspect element 
#class tag is same for all iphone 7 models
flipkart_html.findAll("div",{"class":"_1AtVbE col-12-12"})
```

```python
bigboxes = flipkart_html.findAll("div",{"class":"_1AtVbE col-12-12"})
```
